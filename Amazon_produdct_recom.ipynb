```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#40437991e064dde5625be3b1361c91d8

```

```python
#loading the dataset directly from kaggle without downloading it in local computer
#providing the kaggle key and user name ,by creating the api key in kaggle
import opendatasets as od
od.download('https://www.kaggle.com/datasets/irvifa/amazon-product-reviews')
```

```python
#loading the dataset and naming the column coz they where not named
df=pd.read_csv('/content/amazon-product-reviews/ratings_Electronics.csv',
               names =['user_id','item_id','ratings','timestamp'],
               index_col='user_id',nrows=1000)

'''index_col parameter is used to change one of the columns into the index cox
 2 indexes are not reeeuired to reduces dimensions '''
df.head()
#df.info()
```

```python
#Top 10 Most Reviewed Items
item_counts = df['item_id'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=item_counts.index, y=item_counts.values, palette='viridis')
plt.title('Top 10 Most Reviewed Items')
plt.xlabel('Item ID')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


# Top 10 Items by Average Rating
avg_ratings_per_item = df.groupby('item_id')['ratings'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_ratings_per_item.index, y=avg_ratings_per_item.values, palette='plasma')
plt.title('Top 10 Items by Average Rating')
plt.xlabel('Item ID')
plt.ylabel('Average Rating')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
